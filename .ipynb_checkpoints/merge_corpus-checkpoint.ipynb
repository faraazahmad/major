```python
import pandas as pd
```

```python
names_mixed_corpus = pd.Series([], dtype='string')

for i in range(709):
    try:
        batch = pd.read_csv(f'mixed_corpus/batch_{i}.csv')
        names_mixed_corpus = names_mixed_corpus.append(batch['mixed_corpus'])
        print(f'Added batch {i}')
    except IOError:
        continue
```

```python
index = range(709)
df = pd.DataFrame(names_mixed_corpus, columns=['mixed_corpus'], index=index)
df
```
